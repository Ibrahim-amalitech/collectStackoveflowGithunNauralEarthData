```python
import requests
import pandas as pd
from sqlalchemy import create_engine, text
from sqlalchemy.orm import sessionmaker
from sqlalchemy.exc import IntegrityError
import time

# GitHub API endpoint and authentication
url = ""
auth = ()
# Database connection parameters
server_name = ""
database_name = ""
username = ""
password = ""
connection_string = f"mssql+pyodbc://{username}:{password}@{server_name}/{database_name}?driver=ODBC+Driver+17+for+SQL+Server"

# Create a SQLAlchemy engine and session
engine = create_engine(connection_string)
Session = sessionmaker(bind=engine)
session = Session()

# Function to insert users into the database
def insert_user(user_data):
    try:
        user_columns = ', '.join([f"'{user_data[col]}'" for col in user_data.keys()])
        insert_query = text(f"INSERT INTO Github_Users_500k ({', '.join(user_data.keys())}) VALUES ({user_columns})")
        session.execute(insert_query)
        session.commit()
    except IntegrityError:
        session.rollback()  # User already exists, so we roll back the transaction

# Function to insert login values into usersToBeSearchedFor
def insert_login(login):
    try:
        insert_query = text(f"INSERT INTO usersToBeSearchedFor (login) VALUES ('{login}')")
        session.execute(insert_query)
        session.commit()
    except IntegrityError:
        session.rollback()  # Login already exists, so we roll back the transaction

# Initialize an empty list to store user data
all_users = []

# Function to fetch users in batches of 20
def fetch_users_in_batches():
    # Create a DataFrame from the combined user data
    df = pd.DataFrame(all_users)

    # Iterate through users and insert into the database
    for _, row in df.iterrows():
        username = row['login']

        # Check if the user already exists in Github_Users_500k
        user_query = text(f"SELECT Count(*) AS user_count FROM Github_Users_500k WHERE login = '{username}'")
        user_exists = session.execute(user_query).scalar()

        # Check if the login exists in usersToBeSearchedFor
        user_query = text(f"SELECT Count(*) AS user_count FROM usersToBeSearchedFor WHERE login = '{username}'")
        user_to_search_exists = session.execute(user_query).scalar()

        # Insert into Github_Users_500k if not exists
        if user_exists == 0:
            print(f"{username} will be inserted into Github_Users_500k")
            insert_user(row)
            print(f"{username} inserted into Github_Users_500k")

        # Insert login into usersToBeSearchedFor if not exists
        if user_to_search_exists == 0:
            print(f"{username} will be inserted into usersToBeSearchedFor")
            insert_login(username)
            print(f"{username} inserted into usersToBeSearchedFor")

    # Clear the user data list after insertion
    print(f"{len(all_users)} fetched so far")
    all_users.clear()

# Initial request to get the first page of users
response = requests.get(url, auth=auth)
data = response.json()

# Check if there are more pages of results
while 'next' in response.links:
    next_url = response.links['next']['url']
    response = requests.get(next_url, auth=auth)
    data.extend(response.json())

    # Insert users in batches of 20
    if len(data) >= 20:
        all_users.extend(data[:20])
        fetch_users_in_batches()
        data = data[20:]
        time.sleep(60)  # Pause for 1 minute to avoid rate limiting

# Insert any remaining users
if len(data) > 0:
    all_users.extend(data)
    fetch_users_in_batches()

# Close the database session
session.close()
```

# FETCHING REPOSITORIES FOR GITHUB USERS MATCHED WITH STACKOVERFLOW USERS

```python
from github import Github, GithubException, Auth, UnknownObjectException
import pyodbc
from sqlalchemy import create_engine, text
from sqlalchemy.orm import sessionmaker
import csv
import pandas as pd
import time

# Replace with your GitHub access token
access_token = ''
auth = Auth.Token(access_token)
g = Github(auth=auth)

# # Read the list of names to be fetched from the CSV file
# names_to_fetch = []
# with open('G:\\Shared drives\\IGUS Project\\PowerBI Projects\\usersTobeSearchedForIbrahim.csv', mode='r') as users:
#     csv_reader = csv.DictReader(users)
#     for row in csv_reader:
#         names_to_fetch.append(row['login'])

# Function to get users from the 'login' column of the "usersToBeSearchedFor" table
def get_users_from_table():
    try:
        query = text("SELECT login FROM usersToBeSearchedFor")
        result = session.execute(query).fetchall()
        return [row[0] for row in result]
    except Exception as e:
        print(f"An error occurred while fetching users from the table: {e}")
        return []
    
names_to_fetch = get_users_from_table()

# Connect to SQL Server using SQLAlchemy's create_engine

connection_string = f"mssql+pyodbc://{username}:{password}@{server_name}/{database_name}?driver=ODBC+Driver+17+for+SQL+Server"
engine = create_engine(connection_string)

# Create a session using SQLAlchemy's sessionmaker
Session = sessionmaker(bind=engine)
session = Session()

# Function to fetch repositories for a user and handle rate limit exceeded and server error
def get_repositories(user, column_value):
    try:
        # Fetch repositories for the user
        user_repo_data = []
        for repo in user.get_repos(type='public'):
            try:
                repo_info = {
                    'name': repo.name,
                    'language': repo.language,
                    'languages_url': repo.languages_url,
                    'organization': repo.organization,
                    'owner': f'NamedUser(login="{repo.owner.login}")',
                    'subscribers_count': repo.subscribers_count,
                    'topics': repo.get_topics(),
                    'watchers_count': repo.watchers_count,
                    'full_name': repo.full_name,
                    'tags_url': repo.tags_url,
                    'created_at': repo.created_at,
                    'contributors_url': repo.contributors_url,
                    'collaborators_url': repo.collaborators_url,
                    'has_projects': repo.has_projects
                }
                user_repo_data.append(repo_info)
                print(f"{len(user_repo_data)} repositories fetched so far for {column_value}")

                # Break the loop if user_repo_data has about 4000 repositories
                if len(user_repo_data) >= 4000:
                    break
            except GithubException as e:
                if e.status == 451:
                    print(f"Repository '{repo.full_name}' access blocked. Skipping...")
                elif e.status >= 500:
                    print(f"Server error for user {column_value}. Waiting for 40 minutes...")
                    time.sleep(40 * 60)  # Sleep for 40 minutes (in seconds) before continuing
                    fetch_user_repositories(user, column_value)
                else:
                    raise

        # Convert repo_data list to a DataFrame
        repo_df = pd.DataFrame(user_repo_data)

        # Insert data into SQL Server using pandas
        try:
            repo_df.to_sql('Github_Repositories', con=engine, if_exists='append', index=False)
            print(f"Inserted {len(repo_df)} repositories for {column_value} into SQL Server.")
        except Exception as e:
            # Rollback the transaction if an error occurs
            session.rollback()
            print(f"An error occurred while inserting repositories for {column_value} into SQL Server: {e}")
    except GithubException as e:
        if e.status == 403 and "API rate limit exceeded" in str(e):
            print(f"Rate limit exceeded for user {column_value}. Waiting for 40 minutes...")
            time.sleep(40 * 60)  # Sleep for 40 minutes (in seconds) before continuing
            fetch_user_repositories(user, column_value)
        elif e.status >= 500:
            print(f"Server error for user {column_value}. Waiting for 40 minutes...")
            time.sleep(40 * 60)  # Sleep for 40 minutes (in seconds) before continuing
            fetch_user_repositories(user, column_value)
        else:
            print(f"Error for user {column_value}: {e}. Skipping...")


def fetch_user_repositories():            
    # Iterate through the list of names and fetch repositories
    for column_value in names_to_fetch:
        # Check if 'column_value' is not equal to 'loginName' in the 'owner' column
        query = text(f"SELECT Count(*) AS last_id FROM Github_Repositories WHERE owner = 'NamedUser(login=\"{column_value}\")'")
        result = session.execute(query).first()
        last_id = result[0] if result else 0
        try:
            if not last_id >= 1:
                print(f"{column_value} will be fetched")

                # Fetch repositories for the user
                user = g.get_user(column_value)

                # Call the function to fetch user repositories and handle rate limit exceeded and server error
                get_repositories(user, column_value)
        except UnknownObjectException:
            print(f"User '{column_value}' not found. Skipping...")

    # Commit the remaining collected data to SQL Server
    session.commit()

    # Close the SQLAlchemy session
    session.close()
fetch_user_repositories()
```
